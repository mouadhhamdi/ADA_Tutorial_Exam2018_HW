```python
#We read all the excel file with ExcelFile and select the sheet_name with read_excel
xls = pd.ExcelFile(data_folder + 'communes_pop.xls')
communes_pop = pd.read_excel(xls,header=[0,1,2,3],index=False,skiprows=2,skipfooter=7,sheet_name=0)

#Since we have multiple leveles we delete the unamed ones so that my columns names will make more sens
communes_pop.rename(columns=lambda x: '' if 'Unnamed' in x else x,inplace=True) 
communes_pop.columns=[' '.join(c) for c in communes_pop.columns]
communes_pop.rename(columns={"Composantes de l'évolution de la population Etat de la population au 31 décembre":'Etat de la population au 31 décembre'}
                    ,index=str,inplace=True)
```

```python
#creating new dataframes for our dataframe in order to have more order in our code and also make
#our life easier.

df_cantons = communes_pop[communes_pop.index.str.startswith('-')]
df_communes = communes_pop[communes_pop.index.str.startswith('..')]
df_disrtict = communes_pop[communes_pop.index.str.startswith('>>')]
df_pays = communes_pop[communes_pop.index.str.startswith('Suisse')]
```

```python
'''
Here we used the cut function to create intervalls so that we can count the the communes that have 
a population within that interval and plot the histogram. 
'''
nb_step=1000
cut_df=pd.cut(df_communes['Etat de la population au 31 décembre'],
          bins=np.arange(0,max(df_communes['Etat de la population au 31 décembre']+100),nb_step)).to_frame()
cut_df.rename({'Etat de la population au 31 décembre':'count'},inplace=True,axis=1)

#sort the values 
cut_df=cut_df.sort_values(by=['count'])

#count the number of intervalls
cut_df=cut_df['count'].value_counts().to_frame()

#keep only non values count otherwise is two hard to visualize
cut_df=cut_df[cut_df['count']!=0]
cut_df=cut_df.sort_index()
cut_df.head(2)
```

```python
#only plot the first 100 values, otherwise the plot wont be clear. In case you want to check remove the head(100).
ax = cut_df['count'].plot(kind='bar',rot=90, color="b",figsize=(20,8),x='number of communes',edgecolor = 'black')
ax.set_title('number of commune that have population between interval')
plt.show()
```

```python
# We can also plot a histogram with the count, each bin represent the number of commune that population within 
# that inetrval. 
ax =df_communes['Etat de la population au 31 décembre'].\
    plot(kind='hist',rot=90, color="b",figsize=(20,8),x='number of communes',bins=600,edgecolor = 'black')
ax.set_title('number of commune that have population between interval')
plt.show()
```

```python
#since we are going to have multiple plots we defined a function to show the plot when the arguments chenge.
def plot_(bins,cum):
    ax =df_communes['Etat de la population au 31 décembre'].\
        plot(kind='hist',rot=90, color="b",figsize=(20,8),x='number of communes'
             ,bins=bins,cumulative=cum,edgecolor = 'black')
    ax.set_title('number of commune that have population between interval')
    ax.set_yscale('log')
    plt.show()
```

```python
# small bins size to reproduce the same graph in fig.b
bins = np.arange(0,10000,50) #bins
plot_(bins,False)
```

```python
# small bins size to reproduce the same graph in fig.b
bins = np.arange(0,10000,300) #bins
plot_(bins,False)
```

# Comments:

* In the first plot we have chosen a small size of bins, if we have small bins big counts will be contained in different bin each and this will make the artifacts more clear, it's also because of the log representation of the y-axis
* In the second plot, we choosed bigger bins so that bigger count will kind of some up in one bin. Since small counts are much more important, having big bins will allow them to sum up and be the most important of all bins.

As a conclusion we can say the the frequency increase when the size of bins is big.

```python
#choose a bigger size of bins
bins = np.arange(0,10000,1000) #bins
plot_(bins,-1)
```

# Reading and processing the data 

```python
xls = pd.ExcelFile('data/voters.xls')

voters = pd.read_excel(xls,header=[2],skipfooter=55, sheet_name=0)
cantons = pd.read_csv('data/cantons.csv')

voters.rename(columns={'Unnamed: 0':'cantons'},inplace=True)
voters_df = voters[voters.columns.drop(list(voters.\
                                            filter(regex='Unnamed')))][2:][['cantons','UDC','PS','Participation']]
voters_df=voters_df[~voters_df.cantons.isna()]
#We have replced non null values with 0 since we can not plot the null values and also because we have no 
#0 values in our map.
voters_df=voters_df.fillna(0)
voters_df=voters_df.assign(id_=cantons['Code'].values)
```

```python
state_geo_path = 'data/ch-cantons.topojson.json'
geo_json_data = json.load(open(state_geo_path))
```

# A

```python
map_voters=folium.Map(location=[46.8, 8.33],zoom_start=8)
map_voters.choropleth(geo_data=geo_json_data, data=voters_df,
             columns=['id_', 'UDC'],
             topojson='objects.cantons',
             key_on='feature.id',
             fill_color='YlOrRd',name='Percentage of voters for UDC',legend_name='Percentage of voters for UDC')

folium.LayerControl(collapsed=False).add_to(map_voters)

map_voters
```

* Comments: 
We can see from the map above that the UDC have more electors in switzerland German part than the french or the italian ones.

# B

```python
voters_df=voters_df.assign(Population=cantons['Population'].values)
voters_df['Population']=[int(x.split('[')[0].replace(',','')) if \
                         type(x)!=float else x for x in voters_df['Population']]
voters_df['res_voted']=voters_df['Participation']*0.01*voters_df['Population']*voters_df['UDC']*0.01
```

```python
map_population = folium.Map(location=[46.8, 8.33], zoom_start=8)
map_population.choropleth(geo_data=geo_json_data, data=voters_df[['id_', 'res_voted']],
             columns=['id_', 'res_voted'],
             topojson='objects.cantons',
             key_on='feature.id',
             fill_color='YlOrRd',name='Number of residents who voted UDC',legend_name='Number of residents who voted UDC')
folium.LayerControl(collapsed=False).add_to(map_population)

map_population
```

In this section, we focus on two parties that are representative of the left and the right on the Swiss political spectrum. You will propose a way to visualize their influence over time and for each canton.

For each canton, we define 'right lean' in a certain period as follows:

$$\frac{VoteShare_{UDC} - VoteShare_{PS}}{VoteShare_{UDC} + VoteShare_{PS}}$$  

Visualize the right lean of each canton on the map. What conclusions can you draw this time? Can you observe the [röstigraben](https://en.wikipedia.org/wiki/R%C3%B6stigraben) ?

# A

```python
voters_df=voters_df.assign(Right_lean=(voters_df['UDC']-voters_df['PS'])/(voters_df['UDC']+voters_df['PS']))

map_lean = folium.Map(location=[46.8, 8.33], zoom_start=8)
map_lean.choropleth(geo_data=geo_json_data, data=voters_df[['id_', 'Right_lean']],
             columns=['id_', 'Right_lean'],
             topojson='objects.cantons',
             key_on='feature.id',
             fill_color='YlOrRd',name='Right lean',legend_name='Right lean')

folium.LayerControl(collapsed=False).add_to(map_lean)

map_lean
```
