```python
def tuning():
    # Set the seed
    seed = 12
    # Prepare x_train and y_train to split them into 5 folds
    x_tr = np.array(train.reset_index().drop(['index','target'], axis = 1))
    y_tr = np.array(train['target'])
   
    x_te = np.array(test.reset_index().drop(['index','target'], axis = 1))
    y_te = np.array(test['target'])

    # split data in k fold
    n_estimators = [10, 25, 50, 100]
    max_depth = [2, 4, 10]
    # Table to save the mean accuracy of training the model with the selected features over all folds
    accuracy_tab = [] 
    parameters = []
    
    for estimator in  n_estimators:
        print("estimator: {}".format(estimator))
        for depth in max_depth:
            print("depth: {}".format(depth))
            rf = RandomForestClassifier(n_estimators = estimator,max_depth = depth )
            rf.fit(x_tr, y_tr)
            # get the predicition
            y_pred = rf.predict(x_te)
            
            # Calculate accuracy based on the test sets
            score = accuracy( y_pred,  y_te)
            print("score: {}".format(score))
            accuracy_tab.append(score)
            parameters.append((estimator, depth))
    return accuracy_tab, parameters
```

```python
accuracy_tab, parameters = tuning()
```

```python
index_max = np.argmax(accuracy_tab)
parameters[index_max]
```

```python
def build_k_indices(y, k_fold, seed):
    """build k indices for k-fold."""
    
    num_row = y.shape[0]
    # Set the intervals to split y
    interval = int(num_row / k_fold)
    # We fix the seed to have the same result each call to the function
    np.random.seed(seed)
    indices = np.random.permutation(num_row)
    # Split array of indices into 5 intervals
    k_indices = [indices[k * interval: (k + 1) * interval] for k in range(k_fold)]
    return np.array(k_indices)

def accuracy(pred, y):
    '''
    Calculates the accuracy by comparing the predictions with given test data.
    '''
    N = len(pred)
    count = 0.0
    for i in range(len(pred)):
        if pred[i] == y[i]:
            count += 1
    return count/N

def split_train_test(y,x,k_indices,k): 
    # get k'th subgroup in test, others in train:
        # 1. Set the indices of test and train 
    te_indice = k_indices[k]
    tr_indice = k_indices[~(np.arange(k_indices.shape[0]) == k)]
    tr_indice = tr_indice.reshape(-1)
        # 2. Prepare train and test sets
    y_te = y[te_indice]
    y_tr = y[tr_indice]
    x_te = x[te_indice]
    x_tr = x[tr_indice]
    return y_te,y_tr,x_te,x_tr
```

```python
def train_model(y,x,k_indices, k, estimator, depth):
    """return the accuracy of logistic regression."""
    y_te,y_tr,x_te,x_tr = split_train_test(y,x,k_indices,k)
    # create the logistic model and fitting it with X_train and Y_train 
    rf = RandomForestClassifier(n_estimators = estimator,max_depth = depth )
    rf.fit(x_tr, y_tr)
    # get the predicition
    y_pred = rf.predict(x_te)
    ## Calculate accuracy based on the test sets
    score = accuracy(y_pred, y_te)
    return score
```

```python
def cross_validation():
    # Set the seed
    seed = 12
    num_folds = 6
    # Prepare x_train and y_train to split them into 5 folds
    x = np.array(train.reset_index().drop(['index','target'], axis = 1))
    y =  np.array(train['target'])
    # split data in k fold
    n_estimators = [10, 25, 50, 100]
    max_depth = [2, 4, 10]
    k_indices = build_k_indices(np.array(y), num_folds, seed)

    # Table to save the mean accuracy of training the model with the selected features over all folds
    accuracy = [] 
    parameters = []
    for estimator in  n_estimators:
        print("estimator: {}".format(estimator))
        for depth in max_depth:
            print("depth: {}".format(depth))
            acc_est_dep = []
            for k in range(num_folds):
                score = train_model(y,x,k_indices, k, estimator, depth)
                acc_est_dep.append(score)
            print("mean score: {}".format(np.mean(acc_est_dep)))
            accuracy.append(np.mean(acc_est_dep))
            parameters.append((estimator, depth))
    return accuracy, parameters
```

```python
accuracy_tab, parameters = cross_validation()
```

```python
import numpy as np
index_max = np.argmax(accuracy_tab)

parameters[index_max]
```
